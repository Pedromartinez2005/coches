```python
import pandas as pd
```

```python
#Leemos los csv
coches_francia_df= pd.read_csv('../data/processed/cars_france.csv', index_col= 0)
coches_holanda_df= pd.read_csv('../data/processed/cars_netherlands.csv', index_col= 0)
coches_alemania_df= pd.read_csv('../data/processed/cars_germany.csv', index_col= 0)
coches_españa_df= pd.read_csv('../data/processed/cars_spain.csv' ,index_col= 0)
```

```python
#Cogemos los coches de holanda por separado
df_renault_holanda= coches_holanda_df.loc['RENAULT']
df_seat_holanda= coches_holanda_df.loc['SEAT']
df_opel_holanda= coches_holanda_df.loc['OPEL']
df_hyundai_holanda= coches_holanda_df.loc['HYUNDAI']
df_toyota_holanda= coches_holanda_df.loc['TOYOTA']
df_volkswagen_holanda= coches_holanda_df.loc['VOLKSWAGEN']
```

```python
#Quitamos datos iguales oara que la grafica quedebe bien
df_visualitation_holanda_renault= df_renault_holanda[['kilometros', 'precio']].drop_duplicates(keep= 'first')
df_visualitation_holanda_seat= df_seat_holanda[['kilometros', 'precio']].drop_duplicates(keep= 'first')
df_visualitation_holanda_opel= df_opel_holanda[['kilometros', 'precio']].drop_duplicates(keep= 'first')
df_visualitation_holanda_hyundai= df_hyundai_holanda[['kilometros', 'precio']].drop_duplicates(keep= 'first')
df_visualitation_holanda_toyota= df_toyota_holanda[['kilometros', 'precio']].drop_duplicates(keep= 'first')
df_visualitation_holanda_volkswagen= df_volkswagen_holanda[['kilometros', 'precio']].drop_duplicates(keep= 'first')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(18,9))
ax1= plt.subplot(1,2,1)

sns.regplot(x='kilometros', y='precio', data=df_visualitation_holanda_renault, scatter_kws={'alpha': 0.5, 'color': 'red'}, line_kws={'color': 'red'}, ax= ax1, label= 'RENAULT', ci= None)
sns.regplot(x='kilometros', y='precio', data=df_visualitation_holanda_seat, scatter_kws={'alpha': 0.5, 'color': 'blue'}, line_kws={'color': 'blue'}, ax= ax1, label= 'SEAT', ci= None)
sns.regplot(x='kilometros', y='precio', data=df_visualitation_holanda_opel, scatter_kws={'alpha': 0.5, 'color': 'green'}, line_kws={'color': 'green'}, ax= ax1, label='OPEL', ci= None)

plt.legend()
plt.xlabel('Kilómetros', fontsize=12)
plt.ylabel('Precio', fontsize=12)

ax2= plt.subplot(1,2,2)

sns.regplot(x='kilometros', y='precio', data=df_visualitation_holanda_hyundai, scatter_kws={'alpha': 0.5, 'color': 'purple'}, line_kws={'color': 'purple'}, ax= ax2, label= 'HYUNDAI')
sns.regplot(x='kilometros', y='precio', data=df_visualitation_holanda_toyota, scatter_kws={'alpha': 0.5, 'color': 'orange'}, line_kws={'color': 'orange'}, ax= ax2, label= 'TOYOTA')
sns.regplot(x='kilometros', y='precio', data=df_visualitation_holanda_volkswagen, scatter_kws={'alpha': 0.5, 'color': 'black'}, line_kws={'color': 'black'}, ax= ax2, label='VOLKSWAGEN')

plt.legend()
plt.xlabel('Kilómetros', fontsize=12)
plt.ylabel('Precio', fontsize=12)

plt.suptitle('Relación entre Kilómetros y Precio de Coches-Holanda', fontsize=14)
plt.show()

```
